```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def load_image(file_path):
    try:
        return np.array(Image.open(file_path))
    except Exception as e:
        print(f"Error loading image at {file_path}: {e}")
        return None

def save_image(image_array, output_path):
    Image.fromarray(image_array.astype(np.uint8)).save(output_path)

# Load images using os module
folder = 'Images'
images = [os.path.join(folder, img) for img in os.listdir(folder)]

# Choose the first image for demonstration
input_image_path = images[0]
input_image = load_image(input_image_path)

# Display the original image
plt.imshow(input_image, cmap='gray')
plt.title('Original Image')
plt.show()
```

```python
#Negative transformation
def image_negative(image):
    return 255 - image


negative_image = image_negative(input_image)
plt.imshow(negative_image, cmap='gray')
plt.title('Image Negative')
plt.show()
```

```python
#Log transformation
def log_transformation(image):
    return (np.log1p(image) / np.log1p(255)) * 255

log_transformed_image = log_transformation(input_image)
plt.imshow(log_transformed_image, cmap='gray')
plt.title('Log Transformation')
plt.show()
```

```python
#Inverse Log transformation
def ilog_transformation(image):
    return 255 - (np.log1p(255 - image) / np.log1p(255)) * 255

inverse_log_transformed_image = ilog_transformation(input_image)
plt.imshow(inverse_log_transformed_image, cmap='gray')
plt.title('Inverse Log Transformation')
plt.show()
```

```python
#Gamma correction
def gamma_correction(image, gamma):
    return np.power(image / 255.0, gamma) * 255

gamma_values = [2.5, 1.5,0.5,0 -0.5, -1.5,-2.5,-4]
for gamma in gamma_values:
    gamma_corrected_image = gamma_correction(input_image, gamma)
    plt.imshow(gamma_corrected_image, cmap='gray')
    plt.title(f'Gamma Correction (Gamma={gamma})')
    plt.show()
```
