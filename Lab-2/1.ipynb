USING PILLOW TO GET RGB VALUES

```python
from PIL import ImageColor
ImageColor.getcolor('#ff0000','RGBA')
```

```python
ImageColor.getcolor('red','RGBA')
```

```python
ImageColor.getrgb("hsv(0,100%,100%)")
```

```python
ImageColor.colormap
```

```python
for color in ImageColor.colormap:
    print(f"{color} = {ImageColor.getcolor(color,'RGBA')}")
```

```python
import pandas as pd

Color_Names = list(ImageColor.colormap.keys())
Color_Values = [ImageColor.getcolor(color, 'RGBA') for color in Color_Names]
```

```python
df = pd.DataFrame(Color_Values, columns=['R', 'G', 'B', 'A'], index=Color_Names)
df
```
