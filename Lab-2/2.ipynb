Changing Pixel Colors

```python
from PIL import ImageColor as c 
print(c.getcolor("red","RGBA"))
```

```python
image = i.new("RGBA",(150,150))
color1 = c.getcolor('blue','RGBA')
color2 = c.getcolor('yellow','RGBA')

color = color1
count = 0

for y in range(150): #x,y are rows are columns
    for x in range(150):
        if count == 5:
            color = color1 if color1 != color else color2
            count = 0
        image.putpixel((x,y),color)
        count +=1   
    

image.save('lines.png')         
```

IMAGE BINARIZATION / Converting to black and white


```python
color_image = i.open('image1.jpeg') 
grey_image = color_image.convert("L") #L means Grey-Scale Clean,Without Noise , L stands for LUMA transformation ( Converts RGB to GREY SCALE WITHOUT NOISE)
grey_image.save('grey_image_LUMA.png')
```

```python
color_image = i.open('image1.jpeg') 
random_image = color_image.convert("1") #1 means Grey Scale , But it has Black - Noise  , (Salt[White Noise] and Pepper[Black Noise])
random_image.save('grey_image_noisy.png')
```

```python
color_image = i.open('image1.jpeg') 
random_image = color_image.convert("1",dither=0) #dither = 0 , Removes Noise , Black & White Image 
random_image.save('grey_image_zero_noise.png')
```

Multi Color Image

```python
color_image = i.open('image1.jpeg') 
random_image = color_image.convert("P",palette=i.ADAPTIVE,colors=4)  #chooses just 4 most appropriate colors
#adaptive is given to automate appropriate color selection , we can do it manually too.
random_image = random_image.convert('RGB')
random_image.save('4color.png') 
```
