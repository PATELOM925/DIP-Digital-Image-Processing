##Edge detection by Laplacian filter

````python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

image = Image.open("/content/lena.jpg")

gray_image = image.convert('L')

edge_image = gray_image.filter(ImageFilter.Kernel((3,3),[0,-1,0,-1,4,-1,0,-1,0],2,16))

edge_image_rgb = edge_image.convert('RGB')

plt.imshow(edge_image, cmap=plt.get_cmap('Greys_r'))
````

kernel(size/dimg kernel, [confi], scale. additive)

scale is int value, additive=0

laplasian filter:



```
 0 -1 0
-1 4 -1
 0 -1 0
```
