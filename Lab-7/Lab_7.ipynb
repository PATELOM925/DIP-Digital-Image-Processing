```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
```

```python
image = Image.open("/content/lena.jpg")
```

```python
new_width = image.width // 10
new_height = image.height // 10
resized_image = image.resize((new_width, new_height))
```

```python
gray_image = resized_image.convert('L')
```

```python
# Convert the resized image to grayscale
gray_image = resized_image.convert('L')

# Apply max filter
max_filtered_image = gray_image.filter(ImageFilter.MaxFilter(size=3))

# Apply min filter
min_filtered_image = gray_image.filter(ImageFilter.MinFilter(size=3))

# Apply average filter
average_filtered_image = gray_image.filter(ImageFilter.BoxBlur(radius=1))
```

```python
# Display the images
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
axes[0, 0].imshow(gray_image, cmap='gray')
axes[0, 0].set_title('Original (Resized)')

axes[0, 1].imshow(max_filtered_image, cmap='gray')
axes[0, 1].set_title('Max Filter')

axes[1, 0].imshow(min_filtered_image, cmap='gray')
axes[1, 0].set_title('Min Filter')

axes[1, 1].imshow(average_filtered_image, cmap='gray')
axes[1, 1].set_title('Average Filter')

plt.show()
```
